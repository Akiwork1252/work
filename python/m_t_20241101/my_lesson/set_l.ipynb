```python
n2 = {i for i in range(1, 31) if i % 2 == 0}
n3 = {i for i in range(1, 31) if i % 3 == 0}
n4 = {i for i in range(1, 31) if i % 4 == 0}
n5 = {i for i in range(1, 31) if i % 5 == 0}
n6 = {i for i in range(1, 31) if i % 6 == 0}
n7 = {i for i in range(1, 31) if i % 7 == 0}
n8 = {i for i in range(1, 31) if i % 8 == 0}
n9 = {i for i in range(1, 31) if i % 9 == 0}

n_list = [n2, n3, n4, n5, n6, n7, n8, n9]
n_dict = {}
for idx,n in enumerate(n_list):
    n_dict[idx+2] = n

n_dict
```

```python
for k, v in n_dict.items():
    for l, w in n_dict.items():
        if k >= l:
            pass
        else:
            union = v.union(w)
            intersection = v.intersection(w)
            symmetric_difference = v.symmetric_difference(w)
            l_difference = v.difference(w)
            r_difference = w.difference(v)
            
            print(f'---{k}の倍数と{l}の倍数---')
            print(f'・全ての要素：{len(union)}個{union}')
            print(f'・共通要素：{len(intersection)}個{intersection}')
            print(f'・共通を除く全ての要素：{len(symmetric_difference)}個{symmetric_difference}')
            print(f'・{k}の倍数のみの要素：{len(l_difference)}個{l_difference}')
            print(f'・{l}の倍数のみの要素：{len(r_difference)}個{r_difference}')
    print()
```
